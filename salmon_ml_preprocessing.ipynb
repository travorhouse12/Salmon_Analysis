```python
# Import dependencies
import numpy as np
import pandas as pd
```

```python
# Read in csv file
salmon_df = pd.read_csv('Data/salmon_data.csv')
salmon_df.head()
```

```python
# Delete first column
salmon_df = salmon_df.drop('Unnamed: 0', axis=1)
salmon_df
```

```python
# Drop duplicate rows
salmon_df = salmon_df.drop_duplicates()
salmon_df
```

```python
# Keep only rows that at >= 0 in Effective Catch
salmon_df_filtered = salmon_df[salmon_df['Effective Catch'] >= 0]
salmon_df['Effective Catch'].replace('', np.nan, inplace=True)

# Keep only rows that at >= 0 in Frac Wild
salmon_df_filtered = salmon_df_filtered[salmon_df_filtered['Fracwild'] >= 0]
salmon_df_filtered['Fracwild'].replace('', np.nan, inplace=True)

# Keep only rows that at >= 0 in Frac Wild
salmon_df_filtered = salmon_df_filtered[salmon_df_filtered['Number Of Spawners'] >= 0]
salmon_df_filtered['Number Of Spawners'].replace('', np.nan, inplace=True)

salmon_df_filtered.dropna()

# Reset Index
salmon_df_filtered.reset_index(drop=True, inplace=True)
salmon_df_filtered
```

```python
# Creating Number of Wild Spawners column
salmon_df_filtered["Wild_Spawners"] = salmon_df_filtered["Number Of Spawners"]*salmon_df_filtered["Fracwild"]
# salmon_df_filtered["Wild_Spawners"] = salmon_df_filtered["Number Of Spawners"].multiple(salmon_df_filtered["Fracwild"], axis="index")
salmon_df_filtered.head()
```

```python
# Create a data frame with the last row removed
prev_year = salmon_df_filtered.drop(salmon_df_filtered.index[6666])

# Creating arrays to hold data from the previous years independent variables
prev_spawners = prev_year['Number Of Spawners']
prev_fracwild = prev_year['Fracwild']
prev_eff_catch = prev_year['Effective Catch']
prev_wild_spawners = prev_year['Wild_Spawners']
prev_wild_spawners
#prev_fracwild
#prev_eff_catch
```

```python
# Create a data frame with the last two rows removed
two_yr_prior = salmon_df_filtered.drop(salmon_df_filtered.index[[6665, 6666]])

# Creating arrays to hold data for independent variables from two years prior
two_yr_prior_spawners = two_yr_prior['Number Of Spawners']
two_yr_prior_fracwild = two_yr_prior['Effective Catch']
two_yr_prior_eff_catch = two_yr_prior['Fracwild']
two_yr_prior_wild_spawners = two_yr_prior['Wild_Spawners']
two_yr_prior_wild_spawners
#two_yr_prior_fracwild 
# two_yr_prior_eff_catch
```

```python
# Delete first row
salmon_df_filtered = salmon_df_filtered.drop(salmon_df_filtered.index[0])
salmon_df_filtered.reset_index(drop=True, inplace=True)
salmon_df_filtered.head()
```

```python
# Adding previous year arrays as columns
salmon_df_filtered['Spawners_Prev_Yr'] = prev_spawners.tolist()
salmon_df_filtered['Eff_Catch_Prev_Yr'] = prev_eff_catch.tolist()
salmon_df_filtered['Fracwild_Prev_Yr'] = prev_fracwild.tolist()
salmon_df_filtered['Wild_Spawners_Prev_Yr'] = prev_wild_spawners.tolist()
salmon_df_filtered.head()
```

```python
# Delete first row
salmon_df_filtered = salmon_df_filtered.drop(salmon_df_filtered.index[0])
salmon_df_filtered.reset_index(drop=True, inplace=True)
salmon_df_filtered.head()
```

```python
# Adding two years prior arrays as columns
salmon_df_filtered['Spawners_Two_Yrs_Prior'] = two_yr_prior_spawners.tolist()
salmon_df_filtered['Eff_Catch_Two_Yrs_Prior'] = two_yr_prior_eff_catch.tolist()
salmon_df_filtered['Fracwild_Two_Yrs_Prior'] = two_yr_prior_fracwild.tolist()
salmon_df_filtered['Wild_Spawners_Two_Yrs_Prior'] = two_yr_prior_wild_spawners.tolist()

salmon_df_filtered.head()
```

```python
# Renaming columns
salmon_df_filtered.rename({
    'Brood Year': 'Brood_Year',
    'Nwr Population Name': 'Nwr_Population_Name',
    'Number Of Spawners': 'Number_Of_Spawners',
    'Start Year': 'Start_Year',
    'End Year': 'End_Year',
    'Effective Catch': 'Effective_Catch'},
    inplace=True,
    axis='columns'
)
salmon_df_filtered
```

```python
# Create Pre-Procesing CSV
salmon_df_filtered.to_csv('Data/salmon_preprocessed.csv',index=False)
```
